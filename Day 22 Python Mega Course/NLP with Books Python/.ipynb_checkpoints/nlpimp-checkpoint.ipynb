```python
#Reading the book 
with open("miracle_in_the_andes.txt","r") as file:
    book = file.read()

#book
```

```python
# To calculate how many chapters in this book

#With Strings

book.count("Chapter")
```

```python
#with regex we can count the no of chapters
import re
pattern = re.compile("Chapter [0-9]+")
print("No of Chapters:",re.findall(pattern,book))
print("Total Chapter Length:",len(re.findall(pattern,book)))
```

```python
#To extract sentence which has the word "LOVE" 

pattern = re.compile("[^.]*[^a-zA-Z]+love[^a-zA-Z]+[^.]*.")
res = re.findall(pattern,book)
#print("The Words Found: ",res)
#res
len(res)
```

```python
# Most used Words in this Book

pattern = re.compile("[a-zA-Z]+")
res1 = re.findall(pattern,book)
#res1
len(res1)

```

```python
# to find the words occuring in the book

d = {}

for word in res1:
    if word in d.keys():
        d[word]+=1
    else:
        d[word]=1

# to print the each word count
#d

# to get max count of that word : ie. the word count is used
max(d.values())
```

```python
# to view the data into the list format

d_list = [(value,key) for (key,value) in d.items()]
#d_list
```

```python
# to get that word which has maximum values
sorted(d_list,reverse=True)
```

```python
# Importing nltk and stopwords

import nltk
from nltk.corpus import stopwords
eng_s = stopwords.words("english")
eng_s
```

```python
# to append the all data without stopwords

b_list = []
for count,word in d_list:
    if word not in eng_s:
        b_list.append((count,word))

b_list
        
```
