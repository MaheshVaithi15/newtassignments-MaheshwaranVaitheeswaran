```python
# Spark DataFrame Operations
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('ops').getOrCreate()
print('Spark is Connected')
```

```python
df = spark.read.csv('Datasets/appl_stock.csv',inferSchema=True,header=True)
df.show()
```

```python
# to print schema
df.printSchema()
```

```python
# to filter

df.filter('Close < 500').select(['Open','High','Low']).show()
```

```python
# to get the column which b/w 200 and 300
df.filter((df['Close']>=200) & (df['Close']<=300)).show()
```
