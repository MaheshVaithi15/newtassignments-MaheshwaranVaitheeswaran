```python
import findspark
from pyspark import SparkContext
from pyspark.streaming import StreamingContext
from pyspark.sql import *
from pyspark.sql.functions import *
```

```python
findspark.init()
```

```python
sc = SparkContext()
```

```python
ssc = StreamingContext(sc,10)
```

```python
sqlcontext = SQLContext(sc)
```

```python
socket_stream = ssc.socketTextStream('127.0.0.1',9998)
lines=socket_stream.window(20)
```

```python
from collections import *
fields = ('tag','count')
Tweet = namedtuple('Tweet',fields)
```

```python
lines.flatMap(lambda text : text.split(' ')).filter(lambda word : word.lower().startswith('#')).reduceByKey(lambda a,b:a+b).foreachRDD(lambda rdd : rdd.toDF().sort(desc('count')).limit(10).registerTempTable('Tweets'))
```

```python
ssc.start()
```
